```python
obese_condition = [(df_age_filter['BMI'] < 18.5),
                   (df_age_filter['BMI'] >= 18.5) & (df_age_filter['BMI'] < 25.0),
                   (df_age_filter['BMI'] >= 25.0) & (df_age_filter['BMI'] < 30.0),
                   (df_age_filter['BMI'] > 30.0)]

# 1 - Under Weight
# 2 - Healthy
# 3 - Overweight
# 4 - Obese
obese_value = [1, 2, 3, 4]

df_age_filter['obesity'] = np.select(obese_condition, obese_value, default = 1)
```

### Normalizing data
- Normalize data using MinMaxScaler, because our data are measured in different scales for each column. We need to normalize the data in order to construct and run the models.

### Precision-Recall Curves

---
Precision-Recall Curves is used to evaluate the performance of binary classification algorithms.
Precision is a ratio of the number of true positives divided by the sum of the true positives and false positives. It describes how good a model is at predicting the positive class. Precision is referred to as the positive predictive value.


```python
# Rerun Models and Get the Predict Probabilities
model_svm = svm_clf.fit(X_train, Y_train)
probs_svm = model_svm.predict_proba(X_test)[:, 1]

model_lr = log_reg.fit(X_train, Y_train)
probs_lr = model_lr.predict_proba(X_test)[:, 1] # keep the probabilities for positive outvcomes

model_dt = dt_clf.fit(X_train, Y_train)
probs_dt = model_dt.predict_proba(X_test)[:, 1]

model_rf = rand_forest.fit(X_train,Y_train)
probs_rf = model_rf.predict_proba(X_test)[:, 1]

model_xgb = XGB_Model.fit(X_train, Y_train)
probs_xgb = model_xgb.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import auc, precision_recall_curve

# calculate precision and recall for each model
precision_svm, recall_svm, _ = precision_recall_curve(Y_test, probs_svm)
auc_svm = auc(recall_svm, precision_svm)

precision_lr, recall_lr, _ = precision_recall_curve(Y_test, probs_lr)
auc_lr = auc(recall_lr, precision_lr)

precision_dt, recall_dt, _ = precision_recall_curve(Y_test, probs_dt)
auc_dt = auc(recall_dt, precision_dt)

precision_rf, recall_rf, _  = precision_recall_curve(Y_test, probs_rf)
auc_rf = auc(recall_rf, precision_rf)

precision_xgb, recall_xgb, _ = precision_recall_curve(Y_test, probs_xgb)
auc_xgb = auc(recall_xgb, precision_xgb)
```

```python
# plot the Precision-Recall Curves for each model
plt.figure(figsize=(13, 9))
plt.plot(recall_svm, precision_svm, label='AUC (SVM)')
plt.plot(recall_lr, precision_lr, label='AUC (Logistic Regression)')
plt.plot(recall_dt, precision_dt, label='AUC (Decision Tree)')
plt.plot(recall_rf, precision_rf, label='AUC (Random Forests)')
plt.plot(recall_xgb, precision_xgb, label='AUC (XGBoost)')

plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
plt.show()
```

```python
print(f'AUC of Logistic Regression: {auc_lr:.2f}')
print(f'AUC of SVM: {auc_svm:.2f}')
print(f'AUC of Random Forest: {auc_rf:.2f}')
print(f'AUC of Decision Tree: {auc_dt:.2f}')
print(f'AUC of XGBoost: {auc_xgb:.2f}')
```

- From the Precision-Recall Curves, we can see that the XGBoost Model has the highest AUC values, which is 0.62. Same as the accuracy results, XGBoost have the best fit compared to other models.

```python
#measuring the strength and direction of association between a dichotomous variable obesity and a continuous variable height.
from scipy.stats import pointbiserialr
# Calculate point-biserial correlation
result = pointbiserialr(df_age_filter['obesity2'], df_age_filter['Height'])
print (result)
```

###Justification (Height & Obesity)
####Low correlation but high importance:
The point-biserial correlation coefficient between the 'obesity2' binary variable and the 'Height' continuous variable in our filtered DataFrame df_age_filter is approximately 0.006.
This indicates a very weak correlation between obesity status and height in our dataset.
(indicating no linear relationship between two variables).

And the p-value is approximately 0.654, which is much higher than the common alpha level of 0.05 used to judge statistical significance.
This high p-value suggests that there is not enough evidence to reject the null hypothesis of no correlation between these two variables in the population from which our sample was drawn.
In simpler terms, the relationship between obesity status and height, as observed in our dataset, could very likely be due to chance rather than indicating a true correlation in the broader population.

In interpreting these results, it's important to remember that the lack of a statistically significant correlation does not imply that there is no relationship whatsoever between the two variables; it simply means that any relationship that does exist is not linear or is too weak to be detected with the sample size and variability present in our data.

While height may have a weak correlation with obesity, machine learning models can still find it important for prediction due to non-linear relationships, interaction effects with other features, effective feature engineering during training, model complexity, and potential overfitting to the training data. The models may learn to extract valuable information from height through these mechanisms, leading to its higher importance in prediction despite its low correlation with obesity.

Non-linear relationships: XGBoost can capture non-linear relationships between features and the target variable. Even if the mean heights across obesity categories don't show significant differences, XGBoost might find combinations of features where height plays a crucial role in predicting obesity.
